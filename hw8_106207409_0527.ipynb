```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

### Affine subspace V in R^2.

```python
x = np.array([11,5])
y = np.array([3,10])
d = np.linspace(3,-3,100)
Vx = []
Vy = []

for i in d:
  p = x + np.dot(2*i, y)
  plt.scatter(p[0], p[1], s = 45, c = 'pink')
  Vx.append(p[0])
  Vy.append(p[1])

  plt.axvline(0, c='black')
  plt.axhline(0, c='black')

V = np.array([Vx, Vy]).T
```

```python
A = np.array([[-1,1/4],[0,1]])
p0 = np.array([3,-3])
Ux = []
Uy = []


for i in range(V.shape[0]):
  pp = p0 + np.dot(A, V[i]*1.5)
  plt.scatter(pp[0], pp[1], s = 45, c = 'red')
  Ux.append(pp[0])
  Uy.append(pp[1])

  plt.axvline(0, c='black')
  plt.axhline(0, c='black')

U= np.array([Ux, Uy]).T
```

```python
plt.axvline(0, c='black')
plt.axhline(0, c='black')
plt.scatter(Vx, Vy, s = 50, c = 'pink')
plt.scatter(Ux, Uy, s = 50, c = 'red')
```

我們發現經過Affline mapping後，得到的結果仍然是Affine subspace
